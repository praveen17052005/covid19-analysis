```python
# Full robust loader + EDA block (run as a single cell in Colab)

# Install and imports
!pip install -q plotly seaborn
import pandas as pd, numpy as np, matplotlib.pyplot as plt, seaborn as sns
import plotly.express as px
from google.colab import files
from IPython.display import display, Markdown

plt.rcParams['figure.figsize'] = (12,6)
sns.set(style="whitegrid")

# -------------------------
# Upload file
# -------------------------
print("Please upload your CSV file (e.g., 'covid 19.csv'):")
uploaded = files.upload()
if not uploaded:
    raise SystemExit("No file uploaded - stop and upload your CSV.")
csv_name = list(uploaded.keys())[0]
print("Uploaded file:", csv_name)

# -------------------------
# Load CSV robustly
# -------------------------
df = None
for enc in [None, 'utf-8', 'latin-1']:
    try:
        df = pd.read_csv(csv_name, encoding=enc, low_memory=False)
        print(f"Successfully read with encoding: {enc}")
        break
    except Exception as e:
        print(f"Read attempt with enc={enc} failed: {e}")
if df is None:
    raise SystemExit("Failed to read CSV file.")

display(Markdown("### Preview (first 5 rows)"))
display(df.head())
print("Original columns:", list(df.columns))

# -------------------------
# Normalize column names
# -------------------------
def normalize_colname(c):
    return c.strip().lower().replace(' ', '_').replace('/', '_').replace('-', '_')

orig_cols = list(df.columns)
col_map = {c: normalize_colname(c) for c in orig_cols}
df.rename(columns=col_map, inplace=True)
print("Normalized columns:", list(df.columns))

# -------------------------
# Robust helper to find a column by keywords
# -------------------------
def find_column_by_keywords(df, keywords):
    cols = list(df.columns)
    for kw in keywords:
        # exact match first
        if kw in cols:
            return kw
    # partial match
    for c in cols:
        for kw in keywords:
            if kw in c:
                return c
    return None

# Always define these variables (no NameError later)
date_col = None
country_col = None
cases_col = None
deaths_col = None

# Heuristics: try common names
date_col = find_column_by_keywords(df, ['date', 'report_date', 'day', 'recorded_date'])
country_col = find_column_by_keywords(df, ['country_region', 'country', 'location', 'region', 'country_name'])
cases_col = find_column_by_keywords(df, ['confirmed', 'cases', 'case', 'confirmed_cases', 'new_cases', 'cumulative_cases'])
deaths_col = find_column_by_keywords(df, ['deaths', 'death', 'cumulative_deaths', 'new_deaths'])

print("Detected -> date:", date_col, " country:", country_col, " cases:", cases_col, " deaths:", deaths_col)

# -------------------------
# Parse / clean these columns if present
# -------------------------
# Date parse
if date_col:
    df[date_col] = pd.to_datetime(df[date_col], errors='coerce')
    print(f"Parsed date column '{date_col}', nulls:", df[date_col].isna().sum())
else:
    print("No date column detected.")

# Country normalize
if country_col:
    df[country_col] = df[country_col].astype(str).str.strip()
else:
    print("No country column detected.")

# Numeric conversions - remove commas and coerce
def make_numeric(col):
    df[col] = pd.to_numeric(df[col].astype(str).str.replace(',', '').str.replace('+',''), errors='coerce').fillna(0).astype(int)
    print(f"Converted {col} -> numeric, min={df[col].min()}, max={df[col].max()}")

if cases_col and cases_col in df.columns:
    make_numeric(cases_col)
else:
    # try to find any numeric column that looks like cases
    alt = find_column_by_keywords(df, ['confirmed', 'cases', 'case'])
    if alt:
        cases_col = alt
        make_numeric(cases_col)
    else:
        print("No cases-like column detected.")

if deaths_col and deaths_col in df.columns:
    make_numeric(deaths_col)
else:
    alt2 = find_column_by_keywords(df, ['death', 'deaths'])
    if alt2:
        deaths_col = alt2
        make_numeric(deaths_col)
    else:
        print("No deaths-like column detected.")

# -------------------------
# Basic numeric summary
# -------------------------
display(Markdown("## Numeric columns summary"))
numeric = df.select_dtypes(include=[np.number])
if numeric.shape[1] > 0:
    display(numeric.describe().T)
else:
    print("No numeric columns found.")

# -------------------------
# Aggregate totals by country (if possible)
# -------------------------
country_totals = pd.DataFrame()
if country_col and (cases_col in df.columns or deaths_col in df.columns):
    agg_cols = [c for c in [cases_col, deaths_col] if c in df.columns]
    country_totals = df.groupby(country_col)[agg_cols].sum().reset_index()
    # choose sort key
    sort_key = cases_col if cases_col in country_totals.columns else (deaths_col if deaths_col in country_totals.columns else None)
    if sort_key:
        country_totals = country_totals.sort_values(sort_key, ascending=False)
    display(Markdown("### Country totals (top 10)"))
    display(country_totals.head(10))
else:
    print("Skipping country aggregation: missing country or numeric columns.")

# -------------------------
# Top-N bar chart
# -------------------------
if not country_totals.empty:
    top_n = 10
    top = country_totals.head(top_n).melt(id_vars=country_col, value_vars=[c for c in [cases_col, deaths_col] if c in country_totals.columns])
    fig = px.bar(top, x=country_col, y='value', color='variable', barmode='group',
                 title=f'Top {top_n} Countries')
    fig.update_layout(xaxis_title=None, yaxis_title='Count')
    fig.show()
else:
    print("No country totals to plot.")

# -------------------------
# Time series (global & sample country) if date exists
# -------------------------
if date_col and (cases_col in df.columns or deaths_col in df.columns):
    by_date = df.groupby(date_col)[[c for c in [cases_col, deaths_col] if c in df.columns]].sum().reset_index().sort_values(date_col)
    # create cumulative series (only if values look like daily new; if they are already cumulative you may skip)
    if cases_col in by_date.columns:
        by_date['cum_cases'] = by_date[cases_col].cumsum()
    if deaths_col in by_date.columns:
        by_date['cum_deaths'] = by_date[deaths_col].cumsum()
    # plot
    cols_to_plot = [c for c in ['cum_cases','cum_deaths'] if c in by_date.columns]
    if cols_to_plot:
        fig = px.line(by_date, x=date_col, y=cols_to_plot, title='Global cumulative over time')
        fig.show()
    # sample country plot
    if not country_totals.empty:
        top_country = country_totals.iloc[0][country_col]
        df_c = df[df[country_col] == top_country].groupby(date_col)[[c for c in [cases_col, deaths_col] if c in df.columns]].sum().reset_index().sort_values(date_col)
        if not df_c.empty:
            if cases_col in df_c.columns:
                df_c['cum_cases'] = df_c[cases_col].cumsum()
            if deaths_col in df_c.columns:
                df_c['cum_deaths'] = df_c[deaths_col].cumsum()
            cols = [c for c in ['cum_cases','cum_deaths'] if c in df_c.columns]
            if cols:
                fig = px.line(df_c, x=date_col, y=cols, title=f'Cumulative in {top_country}')
                fig.show()
else:
    print("Skipping time-series: missing date or numeric columns.")

# -------------------------
# Correlation heatmap
# -------------------------
numeric = df.select_dtypes(include=[np.number])
if numeric.shape[1] >= 2:
    corr = numeric.corr()
    display(Markdown("## Correlation matrix"))
    display(corr)
    plt.figure(figsize=(10,8))
    mask = np.triu(np.ones_like(corr, dtype=bool))
    sns.heatmap(corr, mask=mask, annot=True, fmt=".2f", cmap='coolwarm', linewidths=.5)
    plt.show()
else:
    print("Not enough numeric columns for correlation heatmap.")

# -------------------------
# Boxplot: try gender vs numeric, else top countries distribution
# -------------------------
gender_col = find_column_by_keywords(df, ['gender', 'sex'])
target_num = cases_col if cases_col in df.columns else (deaths_col if deaths_col in df.columns else None)

if gender_col and target_num:
    df[gender_col] = df[gender_col].astype(str).str.strip()
    fig = px.box(df, x=gender_col, y=target_num, points='outliers', title=f'{gender_col} vs {target_num}')
    fig.show()
elif not country_totals.empty and target_num:
    top5 = country_totals.head(5)[country_col].tolist()
    subset = df[df[country_col].isin(top5)]
    if not subset.empty:
        fig = px.box(subset, x=country_col, y=target_num, points='outliers', title=f'{target_num} distribution across top 5 countries')
        fig.show()
    else:
        print("Not enough rows to plot boxplot across top countries.")
else:
    print("Skipping boxplot (no gender/target or insufficient data).")

# -------------------------
# Save summary CSV if available
# -------------------------
if not country_totals.empty:
    country_totals.to_csv('covid_country_totals.csv', index=False)
    print("Saved file: covid_country_totals.csv")
else:
    print("No country totals to save.")
```
